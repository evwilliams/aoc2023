# Part 1

```python
def impossible(from_this, subtraction):
    copy = from_this.copy()
    for key, value in copy.items():
        copy[key] = value - (subtraction[key] if key in subtraction else 0)
        if copy[key] < 0:
            return True
    return False


def process(line):
    hypothetical = {
        'red': 12,
        'green': 13,
        'blue': 14,
    }

    min_needed = {
        'red': 0,
        'green': 0,
        'blue': 0,
    }

    splits = line.strip().split(':')
    game = splits[0]
    pulls = splits[1].strip().split(';')
    for pull in pulls:
        colors = pull.strip().split(',')

        counts = {
            'red': 0,
            'green': 0,
            'blue': 0,
        }

        for color in colors:
            info = color.strip().split(' ')
            num = info[0]
            clr = info[1]
            counts[clr] += int(num)

        for key, value in counts.items():
            min_needed[key] = max(counts[key], min_needed[key])

        print(counts)
        if impossible(hypothetical, counts):
            print("Impossible!")
            # return 0

    # This game was possible
    power = 1
    for value in min_needed.values():
        power *= value

    print("Yes!")
    # return int(game.strip(" ").split(' ')[1])
    return power


results = parse(
    filename="day2.txt",
    splitter=line_splitter,
    filterer=lambda s: s is not None and len(s) > 0,
    parser=process)
print(results)
print(len(results))

print(f'Result: {sum(results)}')
```

# Part 2

```python
def get_first_number(line, reverse=False):
    mapping = {
        "one": '1',
        "two": '2',
        "three": '3',
        "four": '4',
        "five": '5',
        "six": '6',
        "seven": '7',
        "eight": '8',
        "nine": '9',
        "zero": '0',
    }
    if reverse:
        line = str_reverse(line)
        mapping = {str_reverse(k): v for k, v in mapping.items()}

    def check(offset, num_chars):
        return mapping.get(line[offset:offset + num_chars])

    for i in range(len(line)):
        if line[i] in mapping.values():
            return line[i]
        for j in range(3, 6):
            if res := check(i, j):
                return res
    return None


def get_number_incl_text(line):
    first, second = get_first_number(line), get_first_number(line, reverse=True)
    return int(first + second)
```

```python
result = parse(
    filename="day1.txt",
    splitter=line_splitter,
    filterer=lambda s: len(s) > 0,
    parser=get_number_incl_text)
print(f'Result: {sum(result)}')
```
